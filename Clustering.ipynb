```python
# Importing required modules
import numpy as np
import pandas as pd
from scipy.spatial.distance import cdist 
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
#Load Data
data = pd.read_csv('/content/Mall_Customers.csv')
data.head()
pca = PCA(2)
```

***Data Processing***

```python
#converting categorical to numeric
data.Gender = pd.Categorical(data.Gender).codes
data.head()
```

```python
# Normalizing the data.
from sklearn.preprocessing import normalize
data_scaled = normalize(data)
data_scaled = pd.DataFrame(data_scaled, columns=data.columns)
data_scaled.head()
```

***K-Means function Implementation***

```python
#Defining our function 
def kmeans(x,k, no_of_iterations):
    idx = np.random.choice(len(x), k, replace=False)
    #Randomly choosing Centroids 
    centroids = x[idx, :]
     
    #finding the distance between centroids and all the data points
    distances = cdist(x, centroids ,'euclidean')
     
    #Centroid with the minimum Distance
    points = np.array([np.argmin(i) for i in distances])
     
    #Repeating the above steps for a defined number of iterations
    for _ in range(no_of_iterations): 
        centroids = []
        for idx in range(k):
            #Updating Centroids by taking mean of Cluster it belongs to
            temp_cent = x[points==idx].mean(axis=0) 
            centroids.append(temp_cent)
 
        centroids = np.vstack(centroids) #Updated Centroids 
         
        distances = cdist(x, centroids ,'euclidean')
        points = np.array([np.argmin(i) for i in distances])
         
    return points 
```

***Clustering using K-Means***

```python
#Transform the data
df = pca.fit_transform(data_scaled)
#Applying Kmeans function
model_kmeans = kmeans(df,3,1000)
#Plot
plt.figure(figsize=(10, 7))  
plt.scatter(data_scaled['Annual Income (k$)'], data_scaled['Spending Score (1-100)'], c=model_kmeans, cmap='viridis') 
```

***Clustering using Agglomerative clustering***

```python
from sklearn.cluster import AgglomerativeClustering
#Applying Agglomerative Clustering function
model_hir = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='complete')
# Fitting data to the model
model_hir.fit_predict(data_scaled)
#Plotting result
plt.figure(figsize=(10, 7))  
plt.scatter(data_scaled['Annual Income (k$)'], data_scaled['Spending Score (1-100)'], c=model_hir.labels_) 
```

***Clustering using Gaussian Mixture***

```python
from sklearn import mixture
#Applying Gaussian Mixture function
model_gm = mixture.GaussianMixture(n_components=3).fit(data_scaled)
#fitting data to the model
labels = model_gm.fit_predict(data_scaled)
#plot
plt.figure(figsize=(10, 7))  
plt.scatter(data_scaled['Annual Income (k$)'], data_scaled['Spending Score (1-100)'], c=labels) 
```

***Clustering using DBSCAN***

```python
from sklearn.cluster import DBSCAN
#Applying DBSCAN function and fitting the data
model_db = DBSCAN(eps=0.2, min_samples=10).fit(data_scaled)
core_samples_mask = np.zeros_like(model_db.labels_, dtype=bool)
core_samples_mask[model_db.core_sample_indices_] = True
labels = model_db.labels_
#Plotting
plt.figure(figsize=(10, 7))  
plt.scatter(data_scaled['Annual Income (k$)'], data_scaled['Spending Score (1-100)'], c=labels, cmap='viridis')
```

```python
from sklearn.metrics import silhouette_score

silhouette_scores = []
k = ['agglomerative','kmeans', 'GM', 'DBSCAN']
silhouette_scores.append(
        silhouette_score(data_scaled, model_hir.fit_predict(data_scaled)))
silhouette_scores.append(
        silhouette_score(data_scaled, model_kmeans))
silhouette_scores.append(
        silhouette_score(data_scaled, model_gm.fit_predict(data_scaled)))
silhouette_scores.append(
        silhouette_score(data_scaled, model_db.fit_predict(data_scaled)))
  
# Plotting a bar graph to compare the results
plt.bar(k, silhouette_scores)
plt.xlabel('Number of clusters', fontsize = 20)
plt.ylabel('S(i)', fontsize = 20)
plt.show()
```
